```python
# Categorical data
```

```python
dfsal = pd.read_csv("./DataSets/hotel_bookings.csv")
```

```python
categorical_col = [col for col in dfsal.columns if dfsal[col].dtype=='object']
```

```python
cat_dfsal = dfsal[categorical_col]
```

```python
#label encoding
```

```python
cat_dfsal["arrival_date_month"] = cat_dfsal["arrival_date_month"].map({"January":1, "February":2, "March":3,"April":4,"May":5, "June":6, "July":7, "August":8, "September":9,"October":10,"November": 11, "December": 12})
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

cat_dfsal["country"] = le.fit_transform(cat_dfsal["country"])
cat_dfsal["hotel"] = le.fit_transform(cat_dfsal["hotel"])
```

```python
cat_dfsal.head(3)
```

```python
#one hot encoding
```

```python
cat_df = pd.get_dummies(data=cat_dfsal, columns=["meal","market_segment","deposit_type","distribution_channel","reserved_room_type","assigned_room_type", "customer_type","reservation_status"])
```

```python
cat_df.drop("reservation_status_date", axis=1, inplace=True)
```

```python
num_df = dfsal.drop(columns = categorical_col, axis=1)
```

```python
final_df = pd.concat([num_df,cat_df], axis=1)
```

```python
# correlation analysis
```

```python
corr_matrix = final_df.corr().abs()
```

```python
threshold = 0.85
```

```python
#remove correlated features
```

```python
import numpy as np

upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
```

```python
to_drop = []
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i,j]) >= threshold:
            colname = corr_matrix.columns[i]
            to_drop.append(colname)
```

```python
final_df.drop(columns=to_drop, inplace=True)
```

```python
final_df["agent"] = final_df["agent"].fillna(0)
```
