```python
from sklearn.datasets import fetch_20newsgroups
twenty_train = fetch_20newsgroups(subset='train', shuffle=True)
```

```python
#svm classifier
```

```python
# Text of the news and goup it belongs to is present in data set
```

```python
#convert text of the news into numeric using Count vectorizer, input it to TFIDF transformer
```

```python
# count vectorizer + tfidf transforer = tdidf vectorizer
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vect = CountVectorizer()
```

```python
x_train_counts = count_vect.fit_transform(twenty_train.data)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
tfidf = TfidfTransformer()
```

```python
x_train_tfidf = tfidf.fit_transform(x_train_counts)
```

```python
from sklearn.svm import LinearSVC
```

```python
svc =LinearSVC(penalty="l2", dual=False, tol=1e-3)
```

```python
svc.fit(x_train_tfidf, twenty_train.target)
```

```python
# Above steps can be simplified by using pipeline of steps
```

```python
from sklearn.pipeline import Pipeline
svc_pipeline = Pipeline(
    [
        ('vect', CountVectorizer()),
        ('tfidf', TfidfTransformer()),
        ('clf', LinearSVC(penalty='l2', dual=False, tol=0.001))
    ]
)
```

```python
# can also be simplified to directly use TfidfVectorizer
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
svc_pipeline = Pipeline(
    [
        ('tfidf', TfidfVectorizer()),
        ('clf', LinearSVC(penalty='l2', dual=False, tol=0.001))
    ]
)
```

```python
svc_pipeline.fit(twenty_train.data, twenty_train.target)
```

```python
# get the test data
```

```python
twenty_test = fetch_20newsgroups(subset='test', shuffle=True)
```

```python
predicted = svc_pipeline.predict(twenty_test.data)
```

```python
# accuracy of the model.
```

```python
from sklearn.metrics import accuracy_score
acc_svm = accuracy_score(twenty_test.target, predicted)
```

```python
acc_svm
```

```python
import pandas as pd
```

```python
mnist_data = pd.read_csv("../DataSets/mnist/train.csv")
```

```python
from sklearn.model_selection import train_test_split
```

```python
features = mnist_data.columns[1:]
```

```python
x = mnist_data[features]
y = mnist_data['label']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x/255, y, test_size=0.1, random_state=0)
```

```python
from sklearn.svm import LinearSVC
```

```python
svm = LinearSVC(penalty='l2',dual=False,tol=1e-5)
```

```python
svm.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = svm.predict(x_test)
acc_svm = accuracy_score(y_test, y_pred)
```

```python
acc_svm
```

```python
# scikit learn provides library to find the best tuning parameters of hyper params
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
penalties = ['l1', 'l2']
tolerances = [1e-3, 1e-4, 1e-5]
```

```python
param_grid = { 'penalty': penalties, 'tol': tolerances}
```

```python
# cv=3, data into 3 sets, 2 training 1 validation
grid_search = GridSearchCV(LinearSVC(dual=False), param_grid, cv=3)
```

```python
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```
